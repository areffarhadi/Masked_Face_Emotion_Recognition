```python
import cv2, os
import matplotlib.pyplot as plt
base_path = "/content/drive/MyDrive/Colab_Notebooks/jaffe/"
new_path = "/content/drive/MyDrive/Colab_Notebooks/jaffe_jpg/"
for infile in os.listdir(base_path):
    print ("file : " + infile)
    read = cv2.imread(base_path + infile)
    outfile =  infile.split('.')[0] +'_'+infile.split('.')[1] +'_'+infile.split('.')[2]+'.jpg'
    outfile2 = cv2.cvtColor(read, cv2.COLOR_BGR2RGB) 
    cv2.imwrite(new_path+outfile,outfile2,[int(cv2.IMWRITE_JPEG_QUALITY), 200])
    print ("file : " + outfile)
    print(new_path+outfile)
```
